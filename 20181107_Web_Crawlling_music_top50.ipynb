```python
import re
```

```python
re.split(('\n\n\r\n\t\t\t\r\n\t\t\t\r\n\t\t\t|\r\n\t\t\n\n'), list_soup_artist[48].get_text())[1]
```

```python
rank = []
song = []
artist = []

for i in range(50):
    rank.append(i+1)
    
    tmp = list_soup_name[i+1].find_all('a')[3].get_text()
    song.append(tmp)
    
    if i<6:
        tmp2 = re.split(('\n\n\r\n\t\t\t\r\n\t\t\t\r\n\t\t\t|\r\n\t\t\n\n'), list_soup_artist[i].get_text())[1]
    elif i==6:
        tmp2 = re.split(('\n|\n'), list_soup_artist_ex[0].get_text())[1]
    elif i>6:
        tmp2 = re.split(('\n\n\r\n\t\t\t\r\n\t\t\t\r\n\t\t\t|\r\n\t\t\n\n'), list_soup_artist[i-1].get_text())[1]
        
    artist.append(tmp2)
```

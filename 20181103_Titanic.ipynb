# 남자/여자 수, 객실 등급별 수

```python
df1 = DataFrame(df, columns=['Pclass','Sex'])
df1
```

```python
df_Male = df1[df1['Sex']=='male']
df_Female = df1[df1['Sex']=='female']
```

```python
# 1,male =>1 : 해당인원
# 2,male =>2 : 해당인원
# 3,male =>3 : 해당인원
df_MPc = {}
df_MPc1 = df_Male[df_Male['Pclass']==1]
df_MPc2 = df_Male[df_Male['Pclass']==2]
df_MPc3 = df_Male[df_Male['Pclass']==3]
df_MPc['first'] = len(df_MPc1)
df_MPc['second'] = len(df_MPc2)
df_MPc['third'] = len(df_MPc3)
df_MPc
```

```python
df_FPc = {}
df_FPc1 = df_Female[df_Female['Pclass']==1]
df_FPc2 = df_Female[df_Female['Pclass']==2]
df_FPc3 = df_Female[df_Female['Pclass']==3]
df_FPc['first'] = len(df_FPc1)
df_FPc['second'] = len(df_FPc2)
df_FPc['third'] = len(df_FPc3)
df_FPc
```

```python
y= np.array(['first','second','third'])
t1= np.array([df_MPc['first'],df_MPc['second'],df_MPc['third']])
t2= np.array([df_FPc['first'],df_FPc['second'],df_FPc['third']])
```

```python
plt.figure(figsize=(8,4))

plt.subplot(121)      # 2x2에 1번째,  subplot(nrows, ncols, plot_number)
plt.bar(y,t1, width=1.0, color=('r','g','b'))
plt.xlabel('Male')

plt.subplot(122)
plt.bar(y,t2, width=1.0, color=('r','g','b'))
plt.xlabel('Female')

plt.show()
```

# 연령대별 인원

```python
df2 = DataFrame(df, columns=['Age'])
df2
```

```python
df_AgeM = {}
df_AgeM['10세 미만'] = 0
df_AgeM['10대'] = 0
df_AgeM['20대'] = 0
df_AgeM['30대'] = 0
df_AgeM['40대'] = 0
df_AgeM['50대'] = 0
df_AgeM['60대'] = 0
df_AgeM['70대'] = 0
df_AgeM['80대'] = 0
df_AgeM['미분류'] = 0
df_AgeM
for i in range(len(df2)):
    if df2['Age'][i] <10:
        df_AgeM['10세 미만'] += 1
    elif df2['Age'][i] <20:
        df_AgeM['10대'] += 1
    elif df2['Age'][i] <30:
        df_AgeM['20대'] += 1
    elif df2['Age'][i] <40:
        df_AgeM['30대'] += 1
    elif df2['Age'][i] <50:
        df_AgeM['40대'] += 1
    elif df2['Age'][i] <60:
        df_AgeM['50대'] += 1
    elif df2['Age'][i] <70:
        df_AgeM['60대'] += 1
    elif df2['Age'][i] <80:
        df_AgeM['70대'] += 1
    elif df2['Age'][i] <90:
        df_AgeM['80대'] += 1
    else:
        df_AgeM['미분류'] += 1
df_AgeM
```

```python
from konlpy.tag import Hannanum
from urllib.request import urlopen
from lxml import html
from collections import Counter

import random
import pytagcloud      # requires Korean font support
import webbrowser

from bs4 import BeautifulSoup 
from urllib.request import urlopen
import pandas as pd
```

```python
rndno = lambda: random.randint(0, 255)
color = lambda: (rndno(), rndno(), rndno())
```

```python
html = 'https://www1.president.go.kr/articles/2029'
response = urlopen(html.format(page=1))

soup = BeautifulSoup(response, "lxml")
soup
```

```python
tmp = soup.find_all('div','text')
tmp
```

```python
text = tmp[1].get_text()
text
```

```python
text = text.replace('\xa0', '')
text
```

```python
def get_tags(text, ntags=50, multiplier=4):
    hannanum = Hannanum()
    nouns = hannanum.nouns(text)
    count = Counter(nouns)
    return [
        {'color' : color(),
         'tag' : n,
         'size' : c*multiplier,
        }
        for n, c in count.most_common(ntags)
    ]
```

```python
def draw_cloud(tags, filename, fontname='Korean', size=(800, 600)):
    pytagcloud.create_tag_image(tags, filename, fontname=fontname, size=size)
    webbrowser.open(filename)
```

```python
tags = get_tags(text)
tags = get_tags(text, ntags=50)
tags[::5]
```

```python
len(tags)
```

```python
result_tags = [] 

# filtering word
number_word = list(str(i) for i in range(10))
circle_word = list('①②③④⑤⑥⑦⑧⑨⑩⑪⑫⑬⑭⑮')
escape_word = list('. , -  ) ( “ \\ \' \" ')
filter_word = number_word + circle_word + escape_word

for tag in tags:
    t_val = tag['tag'] 

    if len(t_val)==1:
        continue
    elif t_val[0] in filter_word or t_val[1] in filter_word:
        continue
    else:
        result_tags.append(tag)
    
len(result_tags)
```
